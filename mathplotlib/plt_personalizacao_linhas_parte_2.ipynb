## Gráfico de linha para apresentação


```python
import matplotlib.pyplot as plt
import numpy as np

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# usando o np where, crie uma lista de marcadores x para o valor mínimo e o para o resto
marcadores = np.where(vendas == np.min(vendas), 'x', '')

plt.plot(meses, vendas)

for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor = 'red'
        tamanho = 80
    else:
        cor = 'blue'
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)

# adicionando uma anotação para mostrar o menor valor em março
plt.annotate('Menor Valor', color='red', xy=(2, 78), xytext=(2.2, 78))



plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

cores = plt.get_cmap('Pastel2').colors
cores_cycler = cycler('color', cores)

plt.rc('axes', prop_cycle = cores_cycler)

plt.figure(figsize=(8, 4))
plt.plot(meses, vendas, marker='o')

plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.box(False)

plt.yticks([])
plt.axvspan(6, 11, color=cores[1], alpha=0.15)
plt.text(6,80, 'Recuperação de crescimento', fontsize=11, color='green', weight='bold')

plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# Ajustando o colormap corretamente
cmap = plt.get_cmap('Pastel2')
cores = [cmap(i) for i in range(cmap.N)]  # Pegando cores disponíveis

cores_cycler = cycler('color', cores)
plt.rc('axes', prop_cycle=cores_cycler)

plt.figure(figsize=(8, 4))
plt.plot(meses, vendas, marker='o', linestyle='-')

# Adicionando os valores acima dos pontos
for i in range(len(meses)):
    plt.text(meses[i], vendas[i] + 2, f'{vendas[i]}', ha='center', fontsize=10, color='black')

plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.box(False)

plt.yticks([])
plt.tick_params(axis='x', length=0)

# Ajustando `axvspan()` para destacar a recuperação
plt.axvspan(meses.index('Jun'), meses.index('Dez'), color=cores[1], alpha=0.15)
plt.text(meses.index('Jun'), 80, 'Recuperação de crescimento', fontsize=11, color='green', weight='bold')

plt.show()
```
