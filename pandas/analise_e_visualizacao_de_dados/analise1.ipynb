```python
import pandas as pd

vendas_df = pd.read_csv(r'../tratamento_de_dados/Contoso - Vendas - 2017.csv', sep=';')
produtos_df = pd.read_csv(r'../tratamento_de_dados/Contoso - Cadastro Produtos.csv', sep=';', encoding='cp860')
clientes_df = pd.read_csv(r'../tratamento_de_dados/Contoso - Clientes.csv', sep=';', encoding='cp860')
lojas_df = pd.read_csv(r'../tratamento_de_dados/Contoso - Lojas.csv', sep=';', encoding='cp860')
promocoes_df = pd.read_csv(r'../tratamento_de_dados/Contoso - Promocoes.csv', sep=';', encoding='cp860')

clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome do Produto']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')


vendas_df = vendas_df.rename(columns={'E-mail': 'Email Cliente'})
display(vendas_df)
```

### Qual cliente que comprou mais vezes?
- Usaremos o método .value_counts() para contar quantas vezes cada valor do Dataframe aparece
- Usaremos o método .plot() para exibir um gráfico

```python
freqClientes = vendas_df['Email Cliente'].value_counts()
display(freqClientes)
freqClientes[:5].plot(figsize=(15,5)) # pegando top5
```

### Qual a Loja que mais vendeu?
- Usaremos o .groupby para agrupar o nosso dataframe, de acordo com o que queremos (somando as quantidades de vendas, por exemplo)

```python
topLojas = vendas_df.groupby(['Nome da Loja']).sum(numeric_only=True)
topLojas = topLojas[['Quantidade Vendida']]
display(topLojas)
```

### Agora precisamos pegar o maior valor. Temos 2 formas:
1. Ordenar o dataframe em ordem decrescente de Quantidade Vendida
- Método .sort_values
2. Pegar o Maior valor diretamente
- Métodos .max() e .idxmax()

```python
topLojas = topLojas.sort_values('Quantidade Vendida', ascending=False)
display(topLojas)
topLojas[:5].plot(figsize=(15,5), kind='barh')
```

```python
maiorValor = topLojas['Quantidade Vendida'].max()
melhorLoja = topLojas['Quantidade Vendida'].idxmax()

print(f'{melhorLoja.strip()} vendeu R$ {maiorValor}')
```

### Qual produto que menos vendeu?
- Já temos uma lista criada para isso, basta verificarmos o final da lista (já que ela está ordenada) ou então usarmos os métodos:
1. min()
2. idxmin()

```python
menorValor = topLojas['Quantidade Vendida'].min() # ou [-1:]
piorLoja = topLojas['Quantidade Vendida'].idxmin()

print(f'{piorLoja.strip()} vendeu R$ {menorValor}')
```
