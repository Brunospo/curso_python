```python
import requests
from bs4 import BeautifulSoup

link = 'https://coinmarketcap.com/'

req = requests.get(link)

site = BeautifulSoup(req.text, 'html.parser')

```

```python
import re

tabela = site.find('tbody')
linhas_da_tabela = tabela.find_all('tr')

for linha in linhas_da_tabela:
  try:
    print(linha.find(class_ = 'coin-item-name').text)
    print(linha.find(string=re.compile(r'\$')))
  except AttributeError:
    pass
```

```python
for linha in linhas_da_tabela:
  texto_linha = linha.get_text(separator=';')
  lista_textos = texto_linha.split(';')
  moeda = lista_textos[1]
  cotacao = lista_textos[3]
  print(lista_textos)
  print(moeda)
  print(cotacao)
```
