### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
```

```python
plt.plot(meses, vendas)
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Editando o Gráfico de Linha

```python
plt.plot(meses, vendas, 'r--o')
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.axis([0,50, 0, max(vendas) + 500])
plt.show()
```

- Dispersão

```python
plt.scatter(meses, vendas)
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.axis([0,50, 0, max(vendas) + 500])
plt.show()
```

- Barras

```python
plt.bar(meses, vendas)
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.axis([0,50, 0, max(vendas) + 500])
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1, figsize=(15,3))
plt.subplot(1,3,1)
plt.plot(meses, vendas, 'r--o')
plt.subplot(1,3,2)
plt.scatter(meses, vendas)
plt.subplot(1,3,3)
plt.bar(meses, vendas)
plt.show()
```
